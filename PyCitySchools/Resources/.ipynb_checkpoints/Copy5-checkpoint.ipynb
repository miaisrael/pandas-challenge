```python
# Import dependencies
import pandas as pd

# Load files
school_data_file = "Resources/schools_complete.csv"
student_data_file = "Resources/students_complete.csv"

# Read and store into Pandas DataFrames
school_data = pd.read_csv(school_data_file)
student_data = pd.read_csv(student_data_file)

# Combine the data into a single dataset.  
combined_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Review
combined_data.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate totals and averages
# Calculate the total number of schools
total_schools= len(school_data["School ID"])

# Calculate total number of students
total_students= len(combined_data["Student ID"])

# Calculate the total budget
total_budget= school_data["budget"].sum()

# Calculate the average math score
avg_math= combined_data["math_score"].mean()

# Calculate the average reading score
avg_reading= combined_data["reading_score"].mean()
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# Create bins to hold passing and failing
math_bins= [0, 69, 100]

# Create labels for bins
math_labels= ["Failed", "Passed"]

# Slice data and place into bins
pd.cut(combined_data["math_score"], math_bins, labels= math_labels)

# Place data series into new column
combined_data["math_status"]= pd.cut(combined_data["math_score"], math_bins, labels= math_labels)

# Calculate the percentage of students passing
math_student_pass_count= combined_data.math_status[combined_data.math_status== "Passed"].count()
math_student_pass_percent= math_student_pass_count/total_students *100
print (math_student_pass_percent)
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
# Create bins to hold passing and failing
reading_bins= [0, 69, 100]

# Create labels for bins
reading_labels= ["Failed", "Passed"]

# Slice data and place into bins
pd.cut(combined_data["reading_score"], reading_bins, labels= reading_labels)

# Place data series into new column
combined_data["reading_status"]= pd.cut(combined_data["reading_score"], reading_bins, labels= reading_labels)

# Calculate the percentage of students passing
reading_student_pass_count= combined_data.reading_status[combined_data.reading_status== "Passed"].count()
reading_student_pass_percent= reading_student_pass_count/total_students *100
print (reading_student_pass_percent)
```

```python
# Check data
combined_data.head()
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)

# Function to pick up both conditions (passing math and reading)
def conditions(combined_data):
    if (combined_data["math_status"] == "Passed") and (combined_data["reading_status"] == "Passed"):
        return 1
    else:
        return 0

# New column for returns
combined_data["overall_status"]= combined_data.apply(conditions, axis=1)

# Total overall passes
overall_passes= combined_data["overall_status"].sum()

# Percentage overall passes
overall_pass_percent= overall_passes/total_students *100
```

```python
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total Number of Schools": [total_schools],
                                 "Total Number of Students": [total_students],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math],
                                 "Average Reading Score": [avg_reading],
                                 "% Students Passing Math": [math_student_pass_percent],
                                 "% Students Passing Reading": [reading_student_pass_percent],
                                 "% Overall Passing": [overall_pass_percent]})

district_summary
```

```python
# Format dataframe

district_summary.style.format({"Total Budget": "${:,.2f}".format})
```

```python
# Function to pick up math status
def math_status(combined_data):
    if (combined_data["math_status"] == "Passed"):
        return 1
    else:
        return 0

# New column for returns
combined_data["math_totals"]= combined_data.apply(math_status, axis=1)
```

```python
# Function to pick up reading status
def reading_status(combined_data):
    if (combined_data["reading_status"] == "Passed"): 
        return 1
    else:
        return 0

# New column for returns
combined_data["reading_totals"]= combined_data.apply(reading_status, axis=1)
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting


```python
combined_data.head()
```

```python
# Create a pandas series for each grade using conditional statement
only_ninth= combined_data[combined_data["grade"] == "9th"]
only_tenth= combined_data[combined_data["grade"] == "10th"]
only_eleventh= combined_data[combined_data["grade"] == "11th"]
only_twelfth= combined_data[combined_data["grade"] == "12th"]
```

```python
# Group each series by school
school_ninth= only_ninth.groupby(["school_name"]).mean()
school_tenth= only_tenth.groupby(["school_name"]).mean()
school_eleventh= only_eleventh.groupby(["school_name"]).mean()
school_twelfth= only_twelfth.groupby(["school_name"]).mean()
```

```python
#Drop not needed columns
drop_nine= school_ninth.drop(["Student ID", "reading_score", "School ID", "size", "budget", "overall_status", \
                  "student_budget", "math_totals", "reading_totals"], axis=1)
drop_ten= school_tenth.drop(["Student ID", "reading_score", "School ID", "size", "budget", "overall_status", \
                  "student_budget", "math_totals", "reading_totals"], axis=1)
drop_eleven= school_eleventh.drop(["Student ID", "reading_score", "School ID", "size", "budget", "overall_status", \
                  "student_budget", "math_totals", "reading_totals"], axis=1)
drop_twelve= school_twelfth.drop(["Student ID", "reading_score", "School ID", "size", "budget", "overall_status", \
                  "student_budget", "math_totals", "reading_totals"], axis=1)
```

```python
# Merge 9th and 10th
math_grade= pd.merge(drop_nine, drop_ten, on="school_name", how="outer")
math_grade.head()
```

```python
# Rename columns
math_grade= math_grade.rename(columns={"math_score_x":"9th", "math_score_y": "10th"})
```

```python
# Merge 10th and 11th
math_grade_middle= pd.merge(math_grade, drop_eleven, on="school_name", how="outer")
math_grade_middle.head()
```

```python
# Final merge with 12th
math_grade_final= pd.merge(math_grade_middle, drop_twelve, on="school_name", how="outer")
math_grade_final.head()
```

```python
# Rename columns and index
math_grade_df= math_grade_final.rename(columns={"math_score_x": "11th", "math_score_y": "12th"})
math_grade_df.index.name= " "
math_grade_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
combined_data.head()
```

```python
# Identify max and min values of per student budget
print(combined_data["student_budget"].max())
print(combined_data["student_budget"].min())
```

```python
# Set bins around this range
budget_range= [0, 579.99, 609.99, 639.99, 675]

# Set bin labels
range_labels= ["<$579", "$580-609", "$610-639", "$640-675"]
```

```python
# Slice and place into bins
pd.cut(combined_data["student_budget"], budget_range, labels= range_labels) 
```

```python
# Add new column with spending ranges
combined_data["spend_range"]= pd.cut(combined_data["student_budget"], budget_range, labels= range_labels) 
combined_data.head()
```

```python
# Group by spending range
spending_group= combined_data.groupby("spend_range")

# Average of needed columns
spending_df=spending_group[["math_score", "reading_score", "math_totals", "reading_totals", "overall_status"]].mean()
```

```python
# Rename columns
spending_df= spending_df.rename(columns= {"math_score": "Average Math Score", \
                                     "reading_score": "Average Reading Score", \
                                     "math_totals": "% Passing Math", \
                                     "reading_totals": "% Passing Reading", \
                                     "overall_status": "% Overall Passing"})

spending_df.head()
```

```python
# Name index
spending_df.index.name="Spending Ranges (Per Student)"
spending_df.head()
```

```python
# Format dataframe
spending_df.style.format({"Average Math Score":"{:.2f}".format,\
                             "Average Reading Score":"{:.2f}".format,\
                             "% Passing Math":"{:.2%}".format, \
                             "% Passing Reading":"{:.2%}".format, \
                             "% Overall Passing":"{:.2%}".format})
```
